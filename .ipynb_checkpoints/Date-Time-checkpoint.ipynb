```python
import pandas as pd
```

```python
df = pd.read_csv('data/ETH_1h.csv')
```

```python
df['Date'] = pd.to_datetime(
    df['Date'],
    format='%Y-%m-%d %I-%p'
)
```

```python
df.loc[0,'Date'].day_name()
```

```python
df['Date'].dt.day_name()
```

```python
df.set_index('Date', inplace = True)
```

```python
df
```

```python
df.resample('W').agg({'Close': 'mean', 'High': 'max', 'Low': 'min', 'Volume': 'sum'}).sort_values(by = 'Date' ,ascending = False)
```

```python
highs = df['High']
```
